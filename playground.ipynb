```python
import re

url_regex = re.compile(r'(https?://[^ ]+m3u8)', re.IGNORECASE)
match = url_regex.search(req.text)
if match:
    url = match.group(0)
    r = requests.get(url)
```

```python
# print(r.content.decode())

qualities = {}
for line in r.content.decode().split('\n'):
    print(line)
    if not line.startswith('#'):
        quality_regex = re.compile(r'\d{3,4}(?=p)', re.IGNORECASE)
        match = quality_regex.search(line)
        if match:
            m = match.group(0)
            qualities[int(m)] = line

print("Qualities available: ", *qualities.keys())
```
